```python
import numpy as np
import sys
sys.path.append("..") 
from Moldata.MolSpace import SpacePCA, SpaceKPCA, SpaceTSNE, SpaceUMAP, SpaceMDS, SpaceSpectral
```

```python
def data_generate() -> np.ndarray:
    '''smiles = np.load("../reactant.npy")
    smiles = np.load("../separator.npy")
    fps = smiles2bits(smiles)'''
    fps = np.load("../fingerpeints.npy")
    return fps

def test_space(data:np.ndarray, name:str='pca') -> None:
    if name == 'pca':
        model = SpacePCA(n_components=None)
        model.plot_pca(data)
    elif name == 'kpca':
        model = SpaceKPCA(n_components=None, kernel='tanimoto')
        model.plot_kpca(data)
    elif name == 'tsne':
        model = SpaceTSNE(metric='jaccard', perplexity=15, learning_rate=50)
    elif name == 'umap':
        model = SpaceUMAP(metric='jaccard')
    elif name == 'mds':
        model = SpaceMDS(metric='minkowski')
        model.plot_shepard(data)
        model.plot_stress(data)
    elif name == 'spectral':
        model = SpaceSpectral(kernel='rbf', metric='jaccard')
        model.plot_spectral(data)
    model.plot_dense(data)
```

```python
test_data = data_generate()
```

```python
test_space(test_data, 'pca')
```

```python
test_space(test_data, 'kpca')
```

```python
test_space(test_data, 'tsne')
```

```python
test_space(test_data, 'umap')
```

```python
test_space(test_data, 'mds')
```

```python
test_space(test_data, 'spectral')
```
