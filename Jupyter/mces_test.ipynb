```python
import sys
sys.path.append("..") 

from Molecule.MoleculeTorch import MoleculeTorch
from Help_Fun.drawer_graph import g2G
from Help_Fun.molpair_mces import MCES

solver_options = {'msg': False,
                  'logPath': '../Temp/ilp_mces.log'}
```

```python
def test_fun(smiles1, smiles2):
    mol1 = MoleculeTorch().from_smiles(smiles1)
    mol2 = MoleculeTorch().from_smiles(smiles2)
    mol1.visualize()
    mol2.visualize()
    mol1.graph_visualize(node_prop='symbol', edge_prop='type', weight=True, figsize=(4, 4))
    mol2.graph_visualize(node_prop='symbol', edge_prop='type', weight=True, figsize=(4, 4))
    mces_result = MCES(smiles1, smiles2, threshold=-1, solver_options=solver_options)
    print(f'Distance between this two molecule: {mces_result[1]}. Time (CPU seconds): {round(mces_result[2], 3)}s')
```

```python
smiles1 = 'CCSC1=C(C=C(C(=C1)OC)CCN)OC'
smiles2 = 'CS/C=C/C(=O)OC[C@H]1CCN2[C@@H]1CCC2'
test_fun(smiles1=smiles1, smiles2=smiles2)
```

```python
smiles1 = 'CC1=C(C=C[N+](=C1CS(=O)(=O)C2=NC3=CC=CC=C3N2)[O-])OCCCOC'
smiles2 = 'CSCCC(C(=O)O)NC(=O)COC1=CC2=C(C=C1)N=C3CCCN3C2=O'
test_fun(smiles1=smiles1, smiles2=smiles2)
```

```python
smiles1 = 'CC1=CC=CC=C1C(C2=CC=CC=C2)O'
smiles2 = 'C1=CC=C(C=C1)COCC2=CC=CC=C2'
test_fun(smiles1=smiles1, smiles2=smiles2)
```
