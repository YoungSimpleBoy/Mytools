```python
import numpy as np
import sys
sys.path.append("..") 
from Molecule.MoleculeAdvance import MoleculeAdvance

file_path_1 = "../SDF_Folder/3d/172330628.sdf"
file_path_2 = "../SDF_Folder/3d/338.sdf"
```

```python
mol = MoleculeAdvance(ignore_H=True).from_file(file_path_1)
mol.visualize(None, width=400, height=400, indices=True)
flipped_mol = MoleculeAdvance(ignore_H=True).from_smiles(mol.flipped_chiral_smiles)
flipped_mol.visualize(None, width=400, height=400, indices=True)
```

```python
print(f'Shape of orthogonal coordinates: {mol._orth_coord().shape}')
print(f'Num of rotatable bonds: {mol.rotatable_bond_num}')
print(f'Molcule volume: {round(mol.molcule_volume, 4)}, Unit: ^3')
print(f'Molcule size (S, L, M) = {mol.molcule_size}')
print(f'Molcule dipole: {mol.molcule_dipole}, Unit: eV')
print(f'TPSA: {mol.tpsa}, Unit: ^2')
print(f'Shape of TPSA contribution: {mol.tpsa_contribs.shape}')
print(f'LogP: {round(mol.logp, 3)}')
print(f"Shape of adjacency matrix: {mol.get_adjacent_bond('length').shape}")
```

```python
print(f'Number of bond angles: {len(mol.get_angle_bond().keys())}')
print(mol.get_angle_bond())
print(mol.get_angle_bond(key='bond'))
print(f'Number of dihedral angle: {len(mol.get_angle_plane().keys())}')
print(mol.get_angle_plane())
print(mol.get_angle_plane(key='angle'))
```

```python
print(mol.angle2atom_b)
print(mol.angle2bond_b)
print(mol.angle2atom_p)
print(mol.angle2bond_p)
```

```python
check_id =45
mol.visualize("../mol.svg", width=500, height=500, indices=True,
              type2d=False,
              highlight_atoms=mol.angle2atom_b[check_id],
              highlight_bonds=mol.angle2bond_b[check_id])
mol.visualize("../mol.svg", width=500, height=500, indices=True,
              type2d=False,
              highlight_atoms=mol.angle2atom_p[check_id],
              highlight_bonds=mol.angle2bond_p[check_id])
```

```python
mol.analyse_energy(verbose=1, n_conformers=200, reset=False)
mol.analyse_angle_bond(verbose=1)
mol.analyse_angle_plane(verbose=1)
```

```python
mol = mol.from_smiles(mol.smiles)
mol.add_H()
flipped_mol.add_H()
mol.optimize3d(n_conformers=200)
flipped_mol.optimize3d(n_conformers=200)
mol.view3D()
flipped_mol.view3D()
```

```python
atom_rep = mol.embedde_atom_rep(nptype=True)
flipped_atom_rep = flipped_mol.embedde_atom_rep(nptype=True)
print(np.max(np.abs(atom_rep - flipped_atom_rep), axis=0))
for i, k in enumerate(mol.embedde_atom_rep().keys()):
    print(f'{k}: {list(mol.embedde_atom_rep().values())[i].shape[1]}')
```

```python
bond_rep = mol.embedde_bond_rep(nptype=True)
flipped_bond_rep = flipped_mol.embedde_bond_rep(nptype=True)
print(np.max(np.abs(bond_rep - flipped_bond_rep), axis=0))
for i, k in enumerate(mol.embedde_bond_rep().keys()):
    print(f'{k}: {list(mol.embedde_bond_rep().values())[i].shape[1]}')
```

```python
import seaborn as sns
angle_rep_b = np.asarray(list(mol.get_angle_bond().values()))
flipped_angle_rep_b = np.asarray(list(flipped_mol.get_angle_bond().values()))
diff = np.abs(angle_rep_b - flipped_angle_rep_b)
g = sns.histplot(data=diff, bins=20, kde=True, fill=True)

sign_indices = [i for i, d in enumerate(diff) if d > 1]
print([round(d, 1) for i, d in enumerate(diff) if i in sign_indices])
check_id = 0
mol.visualize("../mol.svg", width=500, height=500, indices=True,
              type2d=True,
              highlight_atoms=mol.angle2atom_b[sign_indices[check_id]],
              highlight_bonds=mol.angle2bond_b[sign_indices[check_id]])
```

```python
angle_rep_p = np.asarray(list(mol.get_angle_plane().values()))
flipped_angle_rep_p = np.asarray(list(flipped_mol.get_angle_plane().values()))
diff = np.abs(angle_rep_p - flipped_angle_rep_p)
g = sns.histplot(data=diff, bins=20, kde=True, fill=True)

sign_indices = [i for i, d in enumerate(diff) if d > 30]
print([round(d, 1) for i, d in enumerate(diff) if i in sign_indices])
check_id = 0
mol.visualize("../mol.svg", width=500, height=500, indices=True,
              type2d=True,
              highlight_atoms=mol.angle2atom_p[sign_indices[check_id]],
              highlight_bonds=mol.angle2bond_p[sign_indices[check_id]])
```
