```python
import numpy as np
import sys
sys.path.append("..") 
from Moldata.MolCluster import ClusterKMeans, ClusterDBSCAN, ClusterAgglomerative, ClusterGMM, ClusterSpectral, ClusterMeanShift, ClusterOPTICS
```

```python
def data_generate(name:str='iris') -> np.ndarray:
    from sklearn import datasets
    if name == 'iris':
        iris = datasets.load_iris()
        return iris.data[:, :2]
    elif name == 'moon':
        moons, _ = datasets.make_moons(n_samples=300, noise=0.08, random_state=42)
        sphere, _ = datasets.make_circles(n_samples=100, factor=0.5, noise=0.05, random_state=42)
        return np.vstack([moons, sphere*0.5+np.array([1.5, 0.5])])

def test_cluster(data:np.ndarray, name:str='kmeans') -> None:
    if name == 'kmeans':
        model = ClusterKMeans(metric='chebyshev', n_clusters=4).fit(data)
        model.plot_kmeans(data)
    elif name == 'dbscan':
        model = ClusterDBSCAN(eps=0.13, min_samples=6).fit(data)
        model.plot_dbscan(data)
    elif name == 'agglomerative':
        model = ClusterAgglomerative(n_clusters=3).fit(data)
        model.plot_dendrogram(data)
    elif name == 'gmm':
        model = ClusterGMM().fit(data)
        model.plot_gmm(data)
    elif name == 'spectral':
        model = ClusterSpectral(n_clusters=4).fit(data)
        model.plot_spectral(data)
    elif name == 'meanshift':
        model = ClusterMeanShift(bandwidth=0.65).fit(data)
        model.plot_meanshift(data)
    elif name == 'optics':
        model = ClusterOPTICS(xi=0.05, min_samples=10, min_cluster_size=0.05).fit(data)
        model.plot_optics(data)
    model.plot_dense(data)
```

```python
test_data = data_generate('moon')
```

```python
test_cluster(test_data, 'kmeans')
```

```python
test_cluster(test_data, 'dbscan')
```

```python
test_cluster(test_data, 'agglomerative')
```

```python
test_cluster(test_data, 'gmm')
```

```python
test_cluster(test_data, 'spectral')
```

```python
test_cluster(test_data, 'meanshift')
```

```python
test_cluster(test_data, 'optics')
```
